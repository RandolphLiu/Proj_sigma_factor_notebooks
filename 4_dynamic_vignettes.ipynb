## Vignette for different dynamics

The behaviour only include one species i.e. the $\sigma$ factor.
The timecourse is naturally short, but $dt$ is smaller as well.
The line is thicker and the color can be changed.
Pre-stress is not included. The trajectory starts from equilibrium

```julia
include("simu_utils.jl")
pyplot()
```

```julia
# plot function
function plot_vignette(sol, stress_t, max_t, dt; 
        c = :blue, plot_time = 2000.0, max_y = 70.)
    # Select plotting period
    #from = Int64(round((max_t - stress_t) / 2 / dt))
    from = Int64(round(stress_t / dt))
    to = from + Int64(round(plot_time / dt))
    
    # interpret results
    t, σ, ~ = interpret_sol(sol)
    
    # truncate
    t = t[from:to]
    σ = σ[from:to]
    plot(t, σ, legend = :none, linewidth = 2, color = c)
    plot!(grid = :none, framestyle = :box)
    ylims!(0., max_y)
    xlims!(minimum(t), maximum(t))
    plot!(ticks = false)
    plot!(size = (400, 200))
end
```

### No expression

```julia
# reaction definitions
network = new_reaction_system()

# simulation settings
stress_t = 1e3
max_t = 5e4
dt = 0.2

# model parameters
KS = 15.0
rK = 0.1

# simulation
sol = simu_all(network, _β = 50.0, _n = 3.0, _rτ = 5.0, _τ₁ = 10.0,
    _KS = KS, _rK = rK,
    stress_t = stress_t, max_t = max_t, saveat = dt, quiet = true)

# plot
p = plot_vignette(sol, stress_t, max_t, dt, c = :black)
```
