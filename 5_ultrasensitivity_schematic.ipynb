# A schematic figure of biochemical ultrasensitivity

```julia
using Plots
pyplot()    # switch to pyplot backend
```

```julia
# parameters
K_d = 5.    # not to confuse with expression level
β = 1.

# the 1-input, activating hill function
hill(x, K_d, n) = x^n / (K_d^n + x^n)

# plot
x = 0:0.1:10
y = ones(Float64, (3, length(x)))
n_vec = [1, 2, 4]
plot()
for i = 1:3
    y[i, :] .= β * hill.(x, K_d, n_vec[i])
    plot!(x, y[i, :], label = n_vec[i], linewidth = 2)
end
plot!(framestyle = :box, grid = false)
plot!(xlabel = "Regulator", ylabel = "Gene expression")
plot!(legendtitle = "Hill coefficient", legend = :bottomright)
plot!([0., 5., 5.], [0.5, 0.5, 0.], 
    linestyle = :dash, alpha = 0.6,
    color = :blue, label = false)
plot!(xticks = false, yticks = false)
plot!(size = (320, 240))
```
